```python
# heranca_completa.py

class Animal:
    def __init__(self, nome, idade):
        self.nome = nome
        self.idade = idade
        self.vivo = True

    def respirar(self):
        return f"{self.nome} está respirando"

    def comer(self, alimento):
        return f"{self.nome} está comendo {alimento}"

    def morrer(self):
        self.vivo = False
        return f"{self.nome} morreu"

    def __str__(self):
        return f"{self.nome} ({self.idade} anos) - {'Vivo' if self.vivo else 'Morto'}"

class Cachorro(Animal):
    def __init__(self, nome, idade, raca):
        super().__init__(nome, idade)
        self.raca = raca
        self.feliz = True

    def latir(self):
        return "Au au!"

    def brincar(self):
        self.feliz = True
        return f"{self.nome} está brincando e muito feliz!"

    def comer(self, alimento):
        if alimento == "ração":
            return f"{self.nome} está comendo {alimento} com alegria!"
        else:
            return f"{self.nome} recusou {alimento} e quer ração!"

    def __str__(self):
        status = super().__str__()
        return f"{status} - Raça: {self.raca}"

class Gato(Animal):
    def __init__(self, nome, idade, cor):
        super().__init__(nome, idade)
        self.cor = cor
        self.fome = True

    def miar(self):
        return "Miau!"

    def ronronar(self):
        return "Purrrrrr..."

    def comer(self, alimento):
        self.fome = False
        return f"{self.nome} comeu {alimento} elegantemente"

    def dormir(self):
        return f"{self.nome} está tirando uma soneca"

    def __str__(self):
        return f"{super().__str__()} - Cor: {self.cor}"

class AnimalDomestico:
    def __init__(self, dono):
        self.dono = dono
        self.vacinas = []

    def vacinar(self, vacina):
        self.vacinas.append(vacina)
        return f"{self.nome} foi vacinado contra {vacina}"

    def listar_vacinas(self):
        return f"Vacinas: {', '.join(self.vacinas) if self.vacinas else 'Nenhuma vacina'}"

class CachorroDomestico(Cachorro, AnimalDomestico):
    def __init__(self, nome, idade, raca, dono):
        Cachorro.__init__(self, nome, idade, raca)
        AnimalDomestico.__init__(self, dono)

    def __str__(self):
        return f"{super().__str__()} - Dono: {self.dono}"

# Código principal para testar
if __name__ == "__main__":
    print("=== ANIMAL BASE ===")
    animal = Animal("Generic", 2)
    print(animal)
    print(animal.respirar())
    print(animal.comer("grama"))
    print()

    print("=== CACHORRO (Herança Simples) ===")
    rex = Cachorro("Rex", 3, "Labrador")
    print(rex)
    print(rex.respirar())  # Método herdado
    print(rex.latir())     # Método próprio
    print(rex.comer("ração"))  # Método sobrescrito
    print(rex.comer("verdura"))
    print(rex.brincar())
    print()

    print("=== GATO (Herança Simples) ===")
    felix = Gato("Felix", 2, "preto")
    print(felix)
    print(felix.miar())
    print(felix.ronronar())
    print(felix.comer("peixe"))
    print(felix.dormir())
    print()

    print("=== CACHORRO DOMÉSTICO (Herança Múltipla) ===")
    buddy = CachorroDomestico("Buddy", 1, "Golden", "João")
    print(buddy)
    print(buddy.latir())
    print(buddy.vacinar("raiva"))
    print(buddy.vacinar("cinomose"))
    print(buddy.listar_vacinas())
    print()

    print("=== POLIMORFISMO ===")
    animais = [rex, felix, buddy]

    for animal in animais:
        print(f"{animal.nome}: {animal.comer('comida')}")
```
